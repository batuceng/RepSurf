```python
np.random.seed(5)
airplane_ex = np.loadtxt("/mnt/disk5/batu/myprojects/DL2022proj/RepSurf/original_repo/visualization/airplane_0001.txt", delimiter=',')
airplane_ex.shape
inds = np.random.choice(airplane_ex.shape[0], 2048)
airplane_ex = airplane_ex[inds, :]
```
